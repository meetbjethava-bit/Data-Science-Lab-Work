```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
continuous_cols = [
    "Age",
    "Photos",
    "# of Tags",
    "Albums",
    "Posts",
    "Replies",
    "Likes",
    "Events",
    "Friends"
]

# Plot histograms
for col in continuous_cols:
    plt.figure()
    plt.hist(df[col], bins=20)
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.title(f"Histogram of {col}")
    plt.show()
```



---


**Data Visualization Interpretation:**


*   Histograms were plotted for key continuous variables including Age, Photos, # of Tags, Albums, Posts, Replies, Likes, Events, and Friends to examine their distribution patterns.

*   The histogram of Age shows that most users are concentrated between 18 and 25 years old, indicating that the dataset primarily represents young adults. The distribution is slightly right-skewed, with fewer older individuals.

*   The Friends variable is clearly right-skewed. While many users have between 300 and 900 friends, a small number of users have extremely high friend counts (up to 4,999). This suggests the presence of outliers and high variability in social connectivity.
*   The Photos and # of Tags distributions also show strong right-skewness. Most users have moderate activity levels, but a few users have extremely large values (over 10,000 photos or tags), indicating heavy engagement by a small subset of users.
*   The Posts and Replies variables demonstrate similar patterns, where most users have relatively low to moderate activity, while a few users are highly active.
*   Overall, the distributions of engagement-related variables (Photos, Tags, Posts, Replies, Friends) show significant variability and positive skewness, meaning that a small number of users are much more active than the majority.

```python
import numpy as np

numeric_cols = df.select_dtypes(include=np.number).columns
```

```python
for col in numeric_cols:
    if df[col].nunique() <= 5:
        plt.figure()
        df[col].value_counts().plot(kind="bar")
        plt.xlabel(col)
        plt.ylabel("Count")
        plt.title(f"Frequency Plot of {col}")
        plt.show()
```



---


**Frequency Plot (Bar Graph) Interpretation:**

Bar graphs were created for binary and categorical variables such as Gender, Emp, Profile, Cover, Orientation, Relationship, and Edu to examine their distribution.

The bar charts show the frequency of each category (typically 0 and 1 values). Since these variables represent binary attributes, the bar plots clearly display the proportion of users belonging to each category.

For example:


*   The Gender bar plot shows the distribution between the two gender categories.
*   The Relationship variable indicates how many users are in each relationship status category.
*   The Emp, Profile, and Cover variables show whether users have employment information, profile pictures, or cover photos displayed.

The height of each bar represents the count of users in each category, making it easy to compare proportions visually.


---





---


**Insights from Frequency Plots**

The frequency plots help to:

*   Identify dominant categories
*   Compare proportions between binary groups
*   Understand user characteristics
*   Detect imbalances in categorical variables

From the bar graphs, it is evident that some categories (such as having a profile or cover photo) are much more common than others. These plots provide a clearer understanding of the categorical structure of the dataset.

**How Data Visualization Helps**

Data visualization helps to:

*   Identify distribution shape (normal, skewed, symmetric)
*   Detect outliers and extreme values
*   Understand variability within the data
*   Compare activity levels across variables
*   Quickly recognize patterns that are not obvious from tables alone

From the visualizations, it is clear that user engagement metrics are not normally distributed and tend to be positively skewed. This insight is important for understanding user behavior and interpreting the descriptive statistics.
